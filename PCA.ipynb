```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
typologia=pd.read_csv("C:/Users/Maria/Documents/wyniki/aktualne dane/typologia_Bajena1307v4T.csv", index_col=0, sep=';')
df = pd.DataFrame(typologia)
df.head()
```

```python
data = df.drop(labels='typ', axis=0) # axis 1 drops columns, 0 will drop rows that match index value in labels
scaled_data = preprocessing.scale(data.T)
```

```python
pca = PCA()
pca.fit(scaled_data)
pca_data = pca.transform(scaled_data)
pca_data
```

```python
pre_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)
labels = ['PC' + str(x) for x in range(1, len(pre_var)+1)]
```

```python
plt.bar(x=range(1,len(pre_var)+1), height=pre_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Principal Components')
plt.show
```

```python
index=typologia.index
index
```

```python
pca_df = pd.DataFrame(pca_data, index=[list(data)], columns=labels)
pca_df
```

```python
import matplotlib.cm as cm
from matplotlib.lines import Line2D

fig, ax = plt.subplots()


for sample in pca_df.index:
    if "fakt" in str(sample):
        c = 'r'
    if "infpor" in str(sample):
        c = 'g'
    if "konwers" in str(sample):
        c = 'b'
    if "lit" in str(sample):
        c = 'm'
    if "nd" in str(sample):
        c = 'y'
    if "netinterakt" in str(sample):
        c = 'k'
    if "netnieint" in str(sample):
        c = 'coral'
    if "nklas" in str(sample):
        c = 'orange'
    if "publ" in str(sample):
        c = 'aqua'
    if "qmow" in str(sample):
        c = 'tan'
    if "urzed" in str(sample):
        c = 'lime'
    scatter = plt.scatter(pca_df.PC1[sample], pca_df.PC2[sample], c=c, s=2)
    
mark_size=7
    
legend_elements = [Line2D([0], [0], marker='o', color='w', label='fakt',
                          markerfacecolor='r', markersize=mark_size),
                   Line2D([0], [0], marker='o', color='w', label='infpor',
                          markerfacecolor='g', markersize=mark_size),
                   Line2D([0], [0], marker='o', color='w', label='konwers',
                          markerfacecolor='b', markersize=mark_size),
                   Line2D([0], [0], marker='o', color='w', label='lit',
                          markerfacecolor='m', markersize=mark_size),
                   Line2D([0], [0], marker='o', color='w', label='nd',
                          markerfacecolor='y', markersize=mark_size),
                   Line2D([0], [0], marker='o', color='w', label='netinterakt',
                          markerfacecolor='k', markersize=mark_size),
                   Line2D([0], [0], marker='o', color='w', label='netnieint',
                          markerfacecolor='coral', markersize=mark_size),
                   Line2D([0], [0], marker='o', color='w', label='nklas',
                          markerfacecolor='orange', markersize=mark_size),
                   Line2D([0], [0], marker='o', color='w', label='publ',
                          markerfacecolor='aqua', markersize=mark_size),
                   Line2D([0], [0], marker='o', color='w', label='qmow',
                          markerfacecolor='tan', markersize=mark_size),
                   Line2D([0], [0], marker='o', color='w', label='urzed',
                          markerfacecolor='lime', markersize=mark_size),]

legend = ax.legend(handles=legend_elements, loc='upper right')

plt.title('My PCA Graph')
plt.xlabel('PC1 - {0}%'.format(pre_var[0]))
plt.ylabel('PC2 - {0}%'.format(pre_var[1]))
#plt.legend(handles=[legend_elements], loc='lower center', bbox_to_anchor=(1, 0.5))
plt.savefig('wyniki/1607/PCA.png')
plt.show()

```
