```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#loading dataset
typologia=pd.read_csv("C:/Users/Maria/Documents/wyniki/typologia_Bajena6.csv", index_col=0, sep=';')
df1 = pd.DataFrame(typologia)
df1.head()
```

```python
df = df1.iloc[:, :-1]
scaler = MinMaxScaler()
df[df.columns] = scaler.fit_transform(df[df.columns])
df
```

```python
km = KMeans(n_clusters=11)
y_predicted = km.fit_predict(df)
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df.to_csv("C:/Users/Maria/Documents/wyniki/1607/kmeans3_Eder.csv", sep=';')
```

```python
df['typ'] = pd.Series(df1['typ'], index=df.index)
df
```

```python
df3 = df.iloc[:, -2:]
df3
```

```python
clusters = df3.groupby(['cluster','typ']).size()
clusters.to_csv("C:/Users/Maria/Documents/wyniki/1607/kmeans_groupcounts_clustertyp_Eder.csv", sep=';')
clusters
```

```python
clusters = df3.groupby(['typ','cluster']).size()
clusters.to_csv("C:/Users/Maria/Documents/wyniki/1607/kmeans_groupcounts_typcluster_Eder.csv", sep=';')
clusters
```
